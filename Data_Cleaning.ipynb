```python
#Ensuring consistent formatting and unit across the dataset for accurate analysis
```

```python
clean_data['name'] = clean_data['name'].str.lower()
clean_data['host_name'] = clean_data['host_name'].str.lower()
clean_data['neihbourhood_group'] = clean_data['neighbourhood_group'].str.lower()
clean_data['neighbourhood'] = clean_data['neighbourhood'].str.lower()
clean_data['room_type'] = clean_data['room_type'].str.lower()
clean_data.head(10)
```

```python
clean_data['reviews_per_month'] = clean_data['reviews_per_month'].round()
clean_data['reviews_per_month'] = clean_data['reviews_per_month'].astype(int)
clean_data.dtypes
clean_data.head(10)
```

```python
#calculating mean,standard deviation and z-score for each of the data
```

```python
mean = clean_data['price'].mean()
std_dev =clean_data['price'].std()
clean_data['z_score'] = (clean_data['price'] - mean) / std_dev
outliers = clean_data[(clean_data['z_score'] > 3) | (clean_data['z_score'] < -3)]
print("outliers:")
print(outliers)
```
